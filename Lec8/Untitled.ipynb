```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# get the data for plots
DF = pd.read_csv("COVID19-Historical-V2-ST.csv")

x = DF.Date
y = DF.POS_NEW_CP
y1 = DF.POS_7DAYAVG_CP

fig, ax = plt.subplots(figsize=(15,3))

ax.bar(x, y, label="new cases")
ax.set_xlabel("Date")
ax.set_ylabel("New Positive Cases")
ax.set_title("COVID cases in Wisconsin")
ax.plot(x,y1, 'r-', linewidth = 5, label="7-day average")
ax.legend()
start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start, end, 40))
ax.tick_params(axis='x', labelrotation = 45)
print(start)
print(end)
```

```python
x = DF.POS_7DAYAVG_CP
y = DF.DTH_7DAYAVG_CP

xtime = pd.to_datetime(DF.Date)
year = xtime.dt.year

fig, ax = plt.subplots(figsize=(5, 5))
ax.scatter(x, y, s=year-1999, c=year)
ax.set_xlabel("New cases")
ax.set_ylabel("New death")
ax.set_title("Relationship between new cases and new death")

a, b = np.polyfit(x[year==2020], y[year==2020], 1)
ax.plot(x, a*x+b)
a, b = np.polyfit(x[year==2021], y[year==2021], 1)
ax.plot(x, a*x+b)
a, b = np.polyfit(x[year==2022], y[year==2022], 1)
ax.plot(x, a*x+b)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# get the data for plots
DF = pd.read_csv("COVID19-Historical-V2-ST.csv")

x = DF.Date
y1 = DF.POS_7DAYAVG_CP
y2 = DF.NEG_7DAYAVG

fig, ax = plt.subplots(figsize=(15,3))

ax.plot(x, y1, 'g.', label="POSITIVE new cases")
ax.set_xlabel("Date")
ax.set_ylabel("New Positive Cases")
ax.set_title("COVID cases in Wisconsin")
ax.plot(x,y2, 'r:', linewidth = 5, label="NEGATIVE cases")
ax.set_yscale('log')
ax.legend()
start, end = ax.get_xlim()
ax.xaxis.set_ticks(np.arange(start, end, 40))
ax.tick_params(axis='x', labelrotation = 45)
print(start)
print(end)
```

```python

x = DF.Date
y1 = DF.POS_ASN_CP
y2 = DF.POS_BLK_CP
y3 = DF.POS_WHT_CP
y4 = DF.POS_MLTOTH_CP

fig, axes = plt.subplots(2, 2, figsize=(10, 6), sharex = True, sharey = True)
axes[0,0].plot(x, y1, 'k-', label="Asian", linewidth = 2)
axes[0,1].plot(x, y2, 'k-', label="Black", linewidth = 2)
axes[1,0].plot(x, y3, 'k-', label="White", linewidth = 2)
axes[1,1].plot(x, y4, 'k-', label="Others", linewidth = 2)
```

```python
fig, axs = plt.subplot_mosaic([['A', 'C'], ['B', 'C']])
axs['A'].set_title('(a)')
axs['A'].plot(x, y1)
axs['B'].set_title('(b)')
axs['B'].plot(x, y2)
axs['C'].set_title('(c)')
axs['C'].plot(x, y3)
```
