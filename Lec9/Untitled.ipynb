```python
import seaborn as sns
import pandas as pd

pgs = sns.load_dataset("penguins")

pgs
```

```python
sns.set_style('whitegrid')
sns.relplot(data = pgs, x = 'bill_length_mm',
           y = 'bill_depth_mm', hue='species',
           size='body_mass_g', style='sex', kind='line')
```

```python
import seaborn as sns
import pandas as pd

pgs = sns.load_dataset("penguins")

sns.displot(data = pgs, x = "body_mass_g", hue = "island")

sns.displot(data = pgs,
           x = "body_mass_g",
           hue = "island",
            multiple = "stack",
           kind = "kde")
```

```python
# Jointplot

sns.jointplot(data = pgs,
             x = "flipper_length_mm",
             y = "bill_length_mm",
             hue = "species")
```

```python
sns.pairplot(data=pgs, hue = "species")
```

```python
g = sns.PairGrid(data=pgs, hue="species", corner = True)
g.map_lower(sns.kdeplot, hue = None, levels = 4, color="0.4")
g.map_lower(sns.scatterplot, marker="+")
g.map_diag(sns.histplot, kde=True)
g.add_legend()
g.legend.set_bbox_to_anchor((0.6,0.6))
```

```python
# rugplot

sns.rugplot(data=pgs,
            x="flipper_length_mm",
            y="bill_depth_mm",
           hue = "species")
sns.scatterplot(data=pgs,
            x="flipper_length_mm",
            y="bill_depth_mm",
            hue="species")
```

```python
# categorical plot

sns.catplot(data=pgs,
           x="species",
           y="bill_length_mm",
            hue="sex",
           kind="bar")
```

```python
g = sns.lmplot(data=pgs,
               x="bill_length_mm",
               y="bill_depth_mm",
               hue = "species")
g.set_axis_labels("Snoot Length (mm)", "Snooth Depth (mm)")
```

```python
g = sns.lmplot(data=pgs,
               x="bill_length_mm",
               y="bill_depth_mm",
               hue = "species",
              order = 3)
g.set_axis_labels("Snoot Length (mm)", "Snooth Depth (mm)")
```

```python
g = sns.lmplot(data=pgs,
              x = "bill_length_mm",
              y = "bill_depth_mm",
              hue = "species",
              col = "island",
              row = "sex")
```

```python
g = sns.FacetGrid(data = pgs,
                 row = "sex",
                 col = "species",
                  margin_titles = True)
g.map(sns.regplot, "bill_length_mm", "bill_depth_mm", color = "0.4")
```

```python
import pandas as pd
import seaborn as sns
TD = pd.read_csv("TRENDYdata.csv")

TD = TD.loc[TD["tas"]<1000,]

g = sns.lmplot(data = TD,
                x = "tas",
                y = "npp",
               hue = "Model",
              col = "Model",
              col_wrap = 4,
              sharex = False)
```

```python
g = sns.lineplot(data = TD, x = "Year", y = "nbp")
g.set(xlim=(1960, 2010))
```

```python
sns.scatterplot(data=TD,
               x = "rh",
               y = "npp")

import numpy as np
TD["rh_bins"] = pd.cut(x = TD["rh"],
                      bins = np.linspace(40, 70, 4))
sns.catplot(data = TD,
           x = "rh_bins",
           y = "npp",
           kind = "violin")
```
